```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.preprocessing import OneHotEncoder

# Load the dataset into a pandas DataFrame
df = pd.read_csv("newKDDtrain.csv")

# Split the input variables from the target variable
X = df.drop('class', axis=1)
y = df['class']

# One-hot encode the categorical variables
categorical_cols = ['flag']
enc = OneHotEncoder(handle_unknown='ignore')
X_enc = pd.DataFrame(enc.fit_transform(X[categorical_cols]).toarray(),
                     columns=enc.get_feature_names_out(categorical_cols))
X = pd.concat([X.drop(categorical_cols, axis=1), X_enc], axis=1)


# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the classifier on the dataset
clf.fit(X, y)

# Extract the decision tree classification rules
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```
